d. Tăng n, vẽ tiếp và nhận xét

```python
fig, axis = plt.subplots(2, 2, figsize=(10, 10))

for i, ax in enumerate(axis.flat, start=1):
    n = 10 ** i
    x = np.random.normal(1, 1, n)
    y = np.random.normal(2, 1, n)

    ax.scatter(x, y, color="black")
    ax.set_title(f"n points = {n}")







```

a. Tính EX

```python
def f(x):
    return x/2

def g(x):
    return x * f(x)


Ex = integrate.quad(g, 0, 2)
print(Ex)
```
