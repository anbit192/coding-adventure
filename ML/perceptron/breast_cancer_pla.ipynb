```python
import numpy as np
from sklearn.datasets import load_breast_cancer
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split
from sklearn import metrics, linear_model
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
```

# Boob Cancer

- Load data

```python
X, Y = load_breast_cancer(return_X_y=True)
```

```python
def metrics_summary(Y_true, Y_pred):
    acc = metrics.accuracy_score(Y_true, Y_pred)
    recall = metrics.recall_score(Y_true, Y_pred)
    precision = metrics.precision_score(Y_true, Y_pred)

    # confusion_mat = metrics.confusion_matrix(Y_true, Y_pred)

    temp = {"Accuracy":acc, "Recall":recall, "Precision":precision}

    # print(temp)
    return temp
```

## - No PCA

```python
X0_train, X0_test, Y0_train, Y0_test = train_test_split(X, Y, test_size=0.3, random_state=192)
```

```python
model0 = linear_model.Perceptron(max_iter=10000)
model0.fit(X0_train, Y0_train)
```

```python
Y0_predicts = model0.predict(X0_test)
Y0_predicts
```

```python
metrics_summary(Y0_test, Y0_predicts)
```

## - PCA

```python
pca = PCA(n_components=2)
scaler = StandardScaler()
X1 = scaler.fit_transform(X)
X1 = pca.fit_transform(X1)
```

```python
X1_train, X1_test, Y1_train, Y1_test = train_test_split(X1, Y, test_size=0.3, random_state=192)
X1_train
```

```python
model1 = linear_model.Perceptron(max_iter=10000)
model1.fit(X1_train, Y1_train)
```

```python
Y1_predicts = model1.predict(X1_test)
Y1_predicts
```

```python
metrics_summary(Y1_test, Y1_predicts)
```
