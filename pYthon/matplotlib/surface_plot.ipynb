```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import projections
from mpl_toolkits.mplot3d import Axes3D
```

```python
def f(x,y):
    return x**2+y**2
```

```python
x = np.linspace(0,10,11)
y = np.linspace(0,10,11)
xx, yy = coords = np.meshgrid(x,y)
zz = f(xx,yy)
for i in range(2):
    for j in range(11):
        print(coords[i][j])
    print()
```

```python
plt.plot(xx,yy,marker='2',linestyle='none')
plt.show()
```

```python
figure = plt.figure()
ax1 = plt.axes(projection='3d')
ax1.plot_surface(xx,yy,f(xx,yy))
plt.show()
```
