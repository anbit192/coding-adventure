```python
from scipy.integrate import quad, dblquad
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import projections
from mpl_toolkits.mplot3d import Axes3D
```

```python
plt.style.use('dark_background')
```

# Integral 1)

```python
def f1(y,x):
    return (x**3 + x*y)
```

```python
double_integral1 = dblquad(f1, 0,1,lambda x : x**2,lambda x : np.sqrt(x))
print(double_integral1[0])
```

```python
x = np.linspace(0,2,100)
plt.plot(x,x**2,color='red')
plt.plot(x,np.sqrt(x),color='blue')

plt.show()
```

# Integral 2)

```python
def f2(y,x):
    return 1/((x+y)**3)
```

```python
double_integral2 = dblquad(f2,1,2 , 1,lambda x: 3-x )
print(double_integral2[0])
```

```python
x = np.linspace(1,2,100)
plt.plot(x,1+(x-x),color='yellow') #y = 1 
plt.plot(x,3-x,color='red')
plt.show()
```

# Integral 5)

```python
def f5(y,x):
    return x+y
```

```python
double_integral5 = dblquad(f5, 0,1, lambda x : ((1-np.sqrt(x))**2) , lambda x : np.sqrt(1-x**2))
print("Integral of f5: ",double_integral5[0])
```

```python
x = np.linspace(0,1,100) # create domain of x in range (0,1)

plt.plot(x,(1-np.sqrt(x))**2,color='green',label='y = ( 1 - sqrt(x) )^2')
plt.plot(x,np.sqrt(1-x**2),color='purple',label='y = sqrt(1 - x^2)')
plt.plot(x,0+x-x,color='red')
plt.plot(0+x-x,x,color='red')
plt.fill_between(x,(1-np.sqrt(x))**2,np.sqrt(1-x**2))
plt.legend()
plt.show()
```

## Draw surface

```python
def f5_surface(x,y):
    upperbound = np.sqrt(1-x**2)
    lowerbound = (1-np.sqrt(x))**2
    
    return np.where((y<upperbound) & (y>lowerbound), x+y, None)

x_domain = np.linspace(0,1,100)
y_domain = np.linspace(0,1,100)

X , Y = np.meshgrid(x_domain,y_domain)
Z = f5_surface(X,Y)
```

```python
ax5 = plt.axes(projection='3d')
ax5.plot_surface(X,Y,Z)
plt.show()
```

# Integral 6)

```python
def f6(y,x):
    return (x-y)
```

```python
double_integral6 = dblquad(f6, -3,1,lambda x : 2*x-1,lambda x : 2-x**2) 
print(double_integral6[0])
```

```python
x = np.linspace(-3,1,100)
plt.plot(x,2*x-1)
plt.plot(x,2-(x**2))
plt.show()
```

```python
def f6_surface(x,y):
    upperbound = 2-(x**2)
    lowerbound = 2*x-1
    return np.where((y<upperbound) & (y>lowerbound),x-y,None)
```

```python
x = np.linspace(-3,1,100)
y = np.linspace(-7,2,100)

X,Y= np.meshgrid(x,y)
Z = f6_surface(X,Y)
```

```python
ax6 = plt.axes(projection = '3d')
ax6.plot_surface(X,Y,Z)
plt.show()
```

# Integral 7)

```python
def f7(y,x):
    return (x*y)/((x**2)+(y**2))
```

```python
double_integral7 = dblquad(f7,0,1,0,lambda x : 1-x)
print(double_integral7[0])
```
