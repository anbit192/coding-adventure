```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,1,100)
fx = np.sqrt(1-x**2)
gx = (1 - np.sqrt(x))**2
# gx = -np.sqrt(1-x**2)
```

```python
plt.style.use('ggplot')
plt.style.use('dark_background')
figure , axis = plt.subplots(figsize=(5,5))
axis.plot(x,fx,color='yellow',label='fx = sqrt(1-x^2)')
axis.plot(x,gx,color='blue',label='gx = (1-sqrt(x))^2')
axis.fill_between(x,fx, gx)
# axis.plot(x,gx,color='blue',label='gx = - sqrt(1-x^2)')
axis.legend()
```
