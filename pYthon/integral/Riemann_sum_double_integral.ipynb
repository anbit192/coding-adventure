```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x,y):
    upperbound_y = x**2+y**2
    lowerbound_y = np.sqrt(x)+np.sqrt(y)
    
    if ((upperbound_y<1) &(lowerbound_y>1)):
        return x+y
    else:
        return 0
```

```python
def riemann(x_start,x_end,y_start,y_end, partitions):
    
    x = np.linspace(x_start,x_end,partitions) # 0 - 1 => chia ra 100 phan hoach 0 1 2 3 4 5 .. 10
    y = np.linspace(y_start,y_end,partitions) #                                 0 1 2 3 4 5 .. 10
    
    dx = (x_end - x_start) / partitions
    dy = (y_end - y_start) / partitions
    
    dA = dx*dy
    
    riemann_sum = 0
    
    for i in x:
        for j in y:
            z = f(i,j)
            
            riemann_sum += (z * dA)
            
    return riemann_sum
        
```

```python
print(riemann(0,1,0,1,50)) # riemann voi phan hoach = 50
```

# Convergence

```python
riemann_list=[]
x=np.linspace(0,200,1000)
for i in range(10,200):
    riemann_list.append(riemann(0,1,0,1,i))
    
plt.plot(x,3/5+x-x,label='Exact Value')

for i in range(10,200):
    plt.scatter(i,riemann(0,1,0,1,i))
plt.xlabel("partition")
plt.ylabel("values")
plt.legend()
plt.show()
```
