```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import quad
```

## data

```python
start = 0
end = 2
number_of_partition = 100
dx = (end-start)/number_of_partition
f = lambda x : x**2
```

## generate domain of x and y

```python
x_left = np.linspace(start,end-dx,number_of_partition)
y_left = f(x_left)

x_right = np.linspace(dx, end,number_of_partition)
y_right = f(x_right)

x_middle = np.linspace(dx/2, end - (dx/2),number_of_partition)
y_middle = f(x_middle)
```

## Draw

### left riemann

```python
figure1 , axis = plt.subplots(figsize = (5,5))
axis.plot(x_left, y_left, color='blue',label='left riemann')
axis.bar(x_left,y_left,width=(dx),color='purple',align='edge')
axis.legend();
```

### right riemann

```python
figure2 , axis = plt.subplots(figsize=(5,5))
axis.plot(x_right,y_right,color='red',label='riemann right')
axis.bar(x_right,y_right,width=dx,color='blue',align='edge')
axis.legend();
```

### middle riemann

```python
figure3, axis = plt.subplots(figsize=(5,5))
axis.plot(x_middle,y_middle,color='black',label='middle riemann')
axis.bar(x_middle,y_middle,width=dx,color='purple',align='edge')
axis.legend();
```

## calculate

### left riemann

```python
left_sum = 0;
for i in x_left:
    left_sum += (dx * f(i))

print(np.sum(dx*y_left))
print(left_sum)
```

### right riemann

```python
right_sum = 0
for i in x_right:
    right_sum += (f(i)*dx)
    
print(right_sum)
```

### middle riemann

```python
middle_sum = 0
for i in x_middle:
    middle_sum += (f(i)*dx)
    
middle_sum
```
