```python
import pandas as pd
from matplotlib import pyplot as plt
from sklearn import datasets,tree
from sklearn.tree import DecisionTreeClassifier, DecisionTreeRegressor
import numpy as np
```

```python
data = pd.read_csv("C:/Users/ADMIN/Desktop/study/algorithm/decisiontree/datasets/airfoil_self_noise.csv")
heads = ["X0" , "X1", "X2", "X3", "X4"]
x = data.iloc[: , : -1].to_numpy()
y = data.iloc[:, -1].to_numpy()

data.head(10)
```

```python
print(x)
print(y)
```

```python
reg_tree = DecisionTreeRegressor(min_samples_split = 15, min_samples_leaf = 15, max_depth = 3)
model = reg_tree.fit(x, y)

fig = plt.figure(figsize=(25,20))

_ = tree.plot_tree(reg_tree,  filled=True, feature_names = heads)
fig.savefig("regression_plot.png")
plt.show()
```
