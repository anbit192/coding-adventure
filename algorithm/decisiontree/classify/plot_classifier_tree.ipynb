```python
from matplotlib import pyplot as plt
from sklearn import datasets,tree
from sklearn.tree import DecisionTreeClassifier, DecisionTreeRegressor
import numpy as np
import pandas as pd

data = pd.read_csv("C:\\Users\\ADMIN\\Desktop\\study\\algorithm\\decisiontree\\datasets\\iris.csv")
col_names = ["sepal_length", "sepal_width", "petal_length", "petal_width", "type"]

x = data.iloc[:, : -1].to_numpy()
y = data.iloc[:, -1].to_numpy()

data.head(10)


clf_tree = DecisionTreeClassifier(criterion="gini", max_depth = 4, min_samples_split = 10, min_samples_leaf = 5)
model = clf_tree.fit(x, y)

fig = plt.figure(figsize=(25,20))


_ = tree.plot_tree(clf_tree, feature_names=col_names, class_names=y, filled=True)
fig.savefig("flower_classify.png")
```
