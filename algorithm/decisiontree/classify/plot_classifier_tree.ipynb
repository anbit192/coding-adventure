```python
from matplotlib import pyplot as plt
from sklearn import datasets,tree
from sklearn.tree import DecisionTreeClassifier, DecisionTreeRegressor
import numpy as np
import pandas as pd

data = pd.read_csv("C:/Users/ADMIN/Desktop/study/algorithm/decisiontree/datasets/iris.csv")
x = data.iloc[:, :-1]
y = data.iloc[: ,-1]
name_cols = data.columns.tolist()
class_names = y.unique().astype('str')

clf_tree = DecisionTreeClassifier(criterion="gini", max_depth = 6, min_samples_split = 40, min_samples_leaf = 1)
model = clf_tree.fit(x, y)

fig = plt.figure(figsize=(25,20))

_ = tree.plot_tree(clf_tree, feature_names = name_cols, class_names = class_names, filled=True)
fig.savefig("flower_classify.png")
```
