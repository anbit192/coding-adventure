```python
import pandas as pd
from matplotlib import pyplot as plt
from sklearn import datasets,tree
from sklearn.tree import DecisionTreeClassifier, DecisionTreeRegressor
import numpy as np
```

```python
data = pd.read_csv("C:/Users/ADMIN/Desktop/study/algorithm/decisiontree/datasets/winequality-red.csv")
x = data.iloc[: , : -1]
y = data.iloc[:, -1]
name_cols = data.columns.tolist()
name_cols

data
```

```python
clf_tree = DecisionTreeClassifier(criterion="gini", max_depth = 3, min_samples_split = 50, min_samples_leaf = 5)
model = clf_tree.fit(x, y)

fig = plt.figure(figsize=(25,20))
class_names = y.unique().astype('str')
_ = tree.plot_tree(clf_tree, feature_names = name_cols, class_names = class_names ,filled=True)
fig.savefig("wine_classify.png")
```
