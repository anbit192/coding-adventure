# Decision Tree Algorithm
- Giới thiệu: Cây quyết định thuộc thuật toán học có giám sát (supervised learning). Khác với các thuật toán học có giám sát khác, thuật toán này có thế giải quyết các vấn cả bài toán hồi quy (regression) và phân loại (classification).

- `Mục tiêu` của việc sử dụng cây quyết định là tạo ra một mô hình đào tạo có thể sử dụng để dự đoán loại hoặc các giá trị của biến mục tiêu bằng cách học các quy tắc quyết định đơn giản(learning simple decision rules) được suy ra từ dữ liệu trước đó(training data). 
- `Mục tiêu` của việc sử dụng cây quyết định là tìm ra một cây có khả năng dự đoán chính xác cao trên dữ liệu mới chưa biết. Từ đó, ta có thể áp dụng cây quyết định vào nhiều bài toán khác nhau như phân loại spam email, dự đoán giá trị của các bất động sản, phát hiện bệnh ung thư và phân loại ảnh.

- Hai loại cây quyết định: 
    + Categorical Variable Decision Tree: Cây quyết định có biến mục tiêu phân loại thì nó được gọi là  cây quyết định biến phân loại. Ví dụ: phân loại ảnh là mèo hay chó, dự đoán khách hàng có mua sản phẩm hay không.
    + Continuous Variable Decision Tree: Cây quyết định có biến mục tiêu liên tục thì nó được gọi là  Cây quyết định biến liên tục. Ví dụ: dự đoán giá nhà dựa trên diện tích, dự đoán số lượng sản phẩm bán ra.


- Các thuật ngữ trong cây quyết định:
    + Root Node: là nút đầy tiên đại diện cho toàn bộ tập dữ liệu, tiếp tục được chia thành hai hoặc nhiều tập hợp đồng nhất.
    + Leaf Node: Là nút cuối cùng và cây không thể tách biệt thêm được.
    + Splitting: Là quá trình chia các nút thành các nút con theo điều kiện cho trước.
    + Branch/Sub Tree: Một cây được hình thành từ cách tách cây.
    + Pruning: Loại bỏ các nút không mong muốn khỏi cây.
    + Parent/Child node: Nút gốc của cây gọi là parent node, các nút khác gọi là child node.


- Giả sử dựa theo thời tiết mà các bạn nam sẽ quyết định đi đá bóng hay không?


- Dựa theo mô hình trên, ta thấy:

    + Nếu trời nắng, độ ẩm bình thường thì khả năng các bạn nam đi chơi bóng sẽ cao. Còn nếu trời nắng, độ ẩm cao thì khả năng các bạn nam sẽ không đi chơi bóng.

- Xây dựng cây quyết định:
    + Giả sử có bài toán phân loại 2 lớp và mỗi dữ liệu có 2 thuộc tính là x1 và x2.
    
    + Xét điều kiện x1 > 5, chia dữ liệu làm 2 phần, 1 phần thỏa mãn điều kiện và 1 phần không thỏa mãn điều kiện.
    
    + Nếu x1 > 5 tất cả dữ liệu đều thuộc lớp 1 và có thể dùng lớp lá để dự đoán đây là lớp 1.
    + Tiếp tục xét thêm điều kiên x2 > 4.
    
    
    + Nếu điều kiện x2 > 4 đúng thì các dữ liệu thuộc lớp 1, ngược lại các dữ liệu thuộc lớp 0. Do đó 2 node con của node điều kiện trên đều là node lá để cho ra kết quả dự đoán. Kết quả thu được cây quyết định.
    

- Một số thuật toán được sử dụng trong cây quyết định:
    + ID3
    + C4.5: được phát triển từ ID3.
    + CART: Classification And Regression Tree.
    + CHAID: Chi-square Automatic Interaction Detector.
    + MARS: multivariate adaptive regression splines.


- Thuật toán ID3(Iterative Dichotomiser 3): Là thuật toán để xây dựng cây quyết định áp dụng cho các bài toán phân loại bằng cách chọn một thuộc tính tốt nhất để Information Gain(IG) đạt max hoặc Entropy(H) đạt min
