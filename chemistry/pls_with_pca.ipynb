```python

import numpy as np
import pandas as pd
from sklearn.cross_decomposition import PLSRegression
from sklearn.decomposition import PCA
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import train_test_split
    
```

```python
with open("dataCA" ,"r") as f:
    data1 = f.readlines()[0:29]

data1 = [item.strip().replace("\t", " ") for item in data1]
data1 = [" ".join(item.split()).replace(" ", ",") for item in data1]
del data1[0]

temp = []
for l in data1:
    ls = []
    for item in l.split(","):
        ls.append(float(item))
    temp.append(ls)

data1 = temp
del temp

data1 = np.array(data1)
df = pd.DataFrame(data1)
df
```

```python
X = df.iloc[:, :-1]
y = df.iloc[:, -1]
```

```python

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=192)
```

# PLS2

```python
pls2 = PLSRegression(n_components=8)
pls2.fit(X_train, y_train)

y_pred_pls2 = pls2.predict(X_test)

mse_pls2 = mean_squared_error(y_test, y_pred_pls2)
rmse_pls2 = np.sqrt(mse_pls2)
r2_pls2 = r2_score(y_test, y_pred_pls2)

print(f"PLS2 - MSE: {mse_pls2}, RMSE: {rmse_pls2}, R²: {r2_pls2}")
    
```

# PCA + PLS2

```python
pca = PCA(n_components=8)
X_train_pca = pca.fit_transform(X_train)
X_test_pca = pca.transform(X_test)

pls2_pca = PLSRegression(n_components=8)
pls2_pca.fit(X_train_pca, y_train)

y_pred_pca_pls2 = pls2_pca.predict(X_test_pca)

mse_pca_pls2 = mean_squared_error(y_test, y_pred_pca_pls2)
rmse_pca_pls2 = np.sqrt(mse_pca_pls2)
r2_pca_pls2 = r2_score(y_test, y_pred_pca_pls2)

print(f"PCA + PLS2 - MSE: {mse_pca_pls2}, RMSE: {rmse_pca_pls2}, R²: {r2_pca_pls2}")
    
```

```python
results = pd.DataFrame({
    'Method': ['PLS2', 'PCA + PLS2'],
    'MSE': [mse_pls2, mse_pca_pls2],
    'RMSE': [rmse_pls2, rmse_pca_pls2],
    'R2': [r2_pls2, r2_pca_pls2]
})

print(results)
    
```
