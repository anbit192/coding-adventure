```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
img = np.ones((30, 30)) * 120
plt.imshow(img)
img_x = img.shape[0]
img_y = img.shape[1]


```

```python


def convert_cartesian(img):
    new_coords = []



    old_y = 0

    for y in range(-img_y//2, img_y//2):
        old_x = 0

        for x in range(-img_x//2, img_x//2):
            new_coords.append((x, y, old_x, old_y))
            old_x += 1

        old_y += 1

        return new_coords


new_coords = convert_cartesian(img)
new_coords
```

```python
def rotate_vector(angle, point: tuple):
    # origin at (0, 0)
    cart_x = point[0]
    cart_y = point[1]
    old_x = point[2]
    old_y = point[3]

    rotated_x = round(cart_x * math.cos(angle) - cart_y * math.sin(angle))
    rotated_y = round(cart_x * math.sin(angle) + cart_y * math.cos(angle))

    return (rotated_x, rotated_y, old_x, old_y)

   
```

```python
all_x = []
all_y = []
color_point = []

coords_prime = []
for item in new_coords:
    coords_prime.append(rotate_vector(60, item))

for x, y, np_x, np_y in coords_prime:
    all_x.append(x)
    all_y.append(y)
    color_point.append(img[new_coords[np_y][2]][np_x])

coords_prime = np.array(coords_prime)
print(min(all_y))
print(max(all_y))
```

```python
thetas = [0, math.pi/8, math.pi/4, math.pi/3, math.pi/2]

# Generated rotated matrix for all thetas
for theta_in in thetas:
   
    coords_prime = []
    
    print("------------------------------------------")
    print(f"Rotation anticlockwise in degrees = {round((theta_in * 180 / math.pi), 2)}")

    for item in new_coords:
        coords_prime.append(rotate_vector(theta_in, item))
    
    graph_points_all(coords_prime)
```
